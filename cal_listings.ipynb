```python
# Dependencies
import pandas as pd
import requests
import json
import gmaps
import matplotlib.pyplot as plt
import numpy as np
import datetime


# Google developer API key
from config import gkey

# Configure gmaps
gmaps.configure(api_key=gkey)
```

```python
#Store filepath in a variable
file_two = "./listings_calendar.csv"
```

```python
#Read file
file_two_df = pd.read_csv(file_two, encoding="ISO-8859-1")
```

```python
file_two_df.tail(45)
```

```python
file_two_df["date"] = pd.to_datetime(file_two_df["date"])
file_two_df["day_of_week"] = file_two_df["date"].dt.day_name()
file_two_df['price'] = file_two_df['price'].astype(str)

#file_two_df.head()

file_two_df["price"] = file_two_df["price"].apply(lambda x: x.replace("$",""))
file_two_df["price"] = file_two_df["price"].apply(lambda x: x.replace(",",""))

file_two_df['price'] = file_two_df['price'].astype(float)


listings_by_day = file_two_df.groupby("day_of_week").price.mean()
listings_by_day



```

```python
#Visualization #1 property listings by Type

plt.bar(listings_by_day["day_of_week"], listings_by_day["gs"], color="b", align="center")
#print(room_type_percentage_df[0:])
#room_type_viz
plt.title("Number of AirBnB Listings by Type")
plt.xlabel("Listing Types")
```
